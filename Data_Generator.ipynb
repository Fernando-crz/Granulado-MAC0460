```python
import pickle
from matplotlib import pyplot as plt
import torch
from torchvision import datasets, transforms
```

```python
# carregando os datasets utilizados
test_data = datasets.CIFAR100("content/cifar", train=False, transform=transforms.ToTensor(), download=True)
train_data = datasets.CIFAR100("content/cifar", train=True, transform=transforms.ToTensor(), download=True)
```

```python
# visualizando algumas imagens
fig = plt.figure(figsize=(10,9))
rows = 4
columns = 3

for i in range(rows*columns):
  img = test_data[i][0]
  img = torch.einsum("kij->ijk", img)
  fig.add_subplot(columns, rows, i + 1)
  plt.imshow(img)
  plt.title(f"Label: {test_data[i][1]}")
```
